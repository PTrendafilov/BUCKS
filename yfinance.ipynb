# Try the historical data from Yahoo finance

## WHEAT

```python
import yfinance as yf

# Yahoo Finance ticker for CBOT Wheat Futures
ticker = "ZW=F"

# Download monthly data from the start of 2022 to NOW
data = yf.download(
    ticker, 
    start="2022-01-01", 
    end="2024-12-31", 
    interval="1mo"
)

# Print the retrieved DataFrame
print(data)
```

## OIL

```python
import yfinance as yf

# Yahoo Finance ticker for WTI Crude Oil Futures
ticker = "CL=F"

# Download monthly data from the start of 2022 to the end of 2024
data = yf.download(
    ticker, 
    start="2022-01-01", 
    end="2024-12-31", 
    interval="1mo"
)

# Print the retrieved DataFrame
print(data)
```
